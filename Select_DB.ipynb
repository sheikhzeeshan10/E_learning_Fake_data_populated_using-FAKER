```python
#!/usr/bin/python
import sqlite3

def select_DB(table_name):
    conn = sqlite3.connect('olearn.db')

    
    if table_name == 'Employers_Users' :
        for row in conn.execute("SELECT * from Employers_Users"):
            print(row)
            
    elif table_name == 'Course' :
        for row in conn.execute("SELECT * from Course"):
            print(row)
            
    elif table_name == 'Review' :
        for row in conn.execute("SELECT * from Review"):
            print(row)
            
    elif table_name == 'certification_ID' :
        for row in conn.execute("SELECT * from certification_ID"):
            print(row)
            
    elif table_name == 'Ongoing_Training' :
        for row in conn.execute("SELECT * from Ongoing_Training"):
            print(row)
    else:
        print("Error 202!")   

    conn.commit()
    conn.close()


def main():
    db = 'olearn.db'
    table_val = input('Choose the table number you want to insert data:\n 1- User-Employers \n 2- Course \n 3- Review \n 4- Certification ID \n 5- Ongoing Training \n') 
    if table_val == '1':
        table_name = 'Employers_Users' 
    elif table_val == '2':
        table_name = 'Course'
    elif table_val == '3':
        table_name = 'Review'
    elif table_val == '4':
        table_name = 'certification_ID'
    elif table_val == '5':
        table_name = 'Ongoing_Training'
    else:
        print('No such table in databse')
    select_DB(table_name)
        


if __name__=="__main__":
    main()
```
